## Background

There are many factors that help us understand the climate of a country and in this project we are going to look into a few, especially rainfall.

Singapore’s climate is characterised by two main monsoon seasons separated by inter-monsoonal periods.  The **Northeast Monsoon** occurs from December to early March, and the **Southwest Monsoon** from June to September.

The major weather systems affecting Singapore that can lead to heavy rainfall are:

-Monsoon surges, or strong wind episodes in the Northeast Monsoon flow bringing about major rainfall events;

-Sumatra squalls, an organised line of thunderstorms travelling eastward across Singapore, having developed over the island of Sumatra or Straits of Malacca west of us;

-Afternoon and evening thunderstorms caused by strong surface heating and by the sea breeze circulation that develops in the afternoon.

1. Manually calculate mean:

    Write a function that takes in values and returns the mean of the values. Create a list of numbers that you test on your function to check to make sure your function works!
    
    *Note*: Do not use any mean methods built-in to any Python libraries to do this! This should be done without importing any additional libraries.

```python
# Write function that takes in value and return mean of values, using rainy days as an example

def calculate_mean(values):
    # Check if the list is empty, and to Return None if the list is empty
    if len(values) == 0:
        return None  

    # Calculate the sum of all values in the list
    total = sum(values)

    # Calculate the mean by dividing the sum by the number of values
    mean = total / len(values)

    return mean

# Test the function with a sample list of numbers
test_values = [10, 15, 20]
mean_result = calculate_mean(test_values)
print("Mean of test values:", mean_result)
```

2. Manually calculate standard deviation:

    The formula for standard deviation is below:

    $$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

    Where $x_i$ represents each value in the dataset, $\mu$ represents the mean of all values in the dataset and $n$ represents the number of values in the dataset.

    Write a function that takes in values and returns the standard deviation of the values using the formula above. Hint: use the function you wrote above to calculate the mean! Use the list of numbers you created above to test on your function.
    
    *Note*: Do not use any standard deviation methods built-in to any Python libraries to do this! This should be done without importing any additional libraries.

```python
import math

def calculate_standard_deviation(values):
    # Calculate the mean using the function you wrote above
    mean = calculate_mean(values)
    
    # Check if the list is empty
    if len(values) == 0:
        return 0  # Return 0 if the list is empty to avoid division by zero

    # Calculate the sum of squares of differences from the mean
    sum_of_squares = 0
    for value in values:
        sum_of_squares += (value - mean) ** 2

    # Calculate the variance by dividing the sum of squares by the number of values
    variance = sum_of_squares / len(values)

    # Calculate the standard deviation by taking the square root of the variance
    standard_deviation = math.sqrt(variance)
    return standard_deviation

# Test the function with the list of numbers created above
std_dev_result = calculate_standard_deviation(test_values)
print("Standard Deviation:", std_dev_result)
```

```python
# Identifying Outliers for No. of Rainy Days (that which fall outside the 25th and 75th percentile)

Q1 = rain_poi['no_of_rainy_days'].quantile(0.25)
Q3 = rain_poi['no_of_rainy_days'].quantile(0.75)
IQR = Q3 - Q1
filter = (rain_poi['no_of_rainy_days'] >= Q1 - 1.5 * IQR) & (rain_poi['no_of_rainy_days'] <= Q3 + 1.5 * IQR)
outliers = rain_poi[~filter]
#list of outlier months
print("Outlier Months:", outliers['month'])

#count of outlier months
outliers = rain_poi[(rain_poi['no_of_rainy_days'] < Q1) | (rain_poi['no_of_rainy_days'] > Q3)]
number_of_outlier_months = len(outliers)

print("Number of outlier months based on number of rainy days:", number_of_outlier_months)
```
